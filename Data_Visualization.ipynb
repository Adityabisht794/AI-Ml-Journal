# Data visualizatin :
Data visualization is very crucial part of data analysis as it enbales static,animated and interactive visual representaion of insights of data.  

### Libraries Used:
1: Matplotlib.  
2: Seaborn.


## Matplotlib :

```python
## Importing matplotlib library:

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,9,16,25]

## Plotting the data:
print("Plotting the data using matplotlib:")

## Custom line style:
plt.plot(x, y,color='red',linestyle='--',marker='o')

## Adding title and labels:
plt.title("Sample Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
## Plotting multiple lines:

print("Plotting multiple lines:")

x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]

## Plotting multiple lines with different styles:

plt.plot(x, y, label='Line 1', color='blue', marker='*')
plt.plot(x, z, label='Line 2', color='green', marker='o')
plt.title("Multiple Lines Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
## Plotting multiple graph:

print("Plotting multiple lines:")
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]

plt.Figure(figsize=(8, 5))

plt.subplot(2, 2, 1)
plt.plot(x, y, label='Line:1', color='blue', marker='o')
plt.title("Graph:1")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(y, z, label='Line:2', color='green', marker='o')
plt.title("Graph:2")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(y, z, label="Line:3", color="red", marker="o")
plt.title("Graph:3")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, x, label="Line:4", color="yellow", marker="o")
plt.title("Graph:4")
plt.legend()
```

```python
## Plotting Bar Graph:

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

print("Plotting Bar Graph:")

plt.bar(categories, values, color='black')
plt.title("Bar Graph Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
## Plotion histogram:

data= [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 6, 6, 7, 8]
print("Plotting Histogram:")
plt.hist(data, bins=4, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
## Scatter Plot:

import pandas as pd

df=pd.read_csv("placement.csv")
print("Plotting Scatter Plot:")
plt.scatter(df['IQ'], df['CGPA'], color='blue', marker='.')
plt.title("Scatter Plot :")
plt.xlabel("Serial Number")
plt.ylabel("Salary")
plt.show()
```

```python
## IQ trend:

iq_trend= df.groupby('College_ID')['CGPA'].sum().reset_index()
plt.plot(iq_trend['College_ID'], iq_trend['CGPA'], color='red', marker='o')
print(iq_trend) 
```
