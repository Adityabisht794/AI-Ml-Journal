## Handling of Missing Values:


```python
## creation of small dataframe for handling missing values:
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David' , 'Eve'],
    'Age': [25, None, 30, 22 ,None],
    'Salary': [None, 60000, None, 45000 ,None]
}
df = pd.DataFrame(data)
print("Data-Frame:\n")
df
```

```python
### use of isnull() :
df.isnull()
```

```python
## count of missing values in each column:
df.isnull().sum()
```

```python
## Handling of Missing Values with 0: :
df.fillna(0,inplace=True)
print("Data-frame after filling missing values with 0:\n") 
df
```

```python
## Handling of Missing Values with mean: :
df["Salary"] = df["Salary"].replace(0, df["Salary"].mean())
df["Age"] = df["Age"].replace(0, df["Age"].mean())
print("Data-frame after filling missing values with mean:\n")
df
```

```python
## Changing name of a column:

df.rename(columns={'Name': 'Identity'},inplace=True)
df
```

```python
## Changing data type of a column:

df['Age'] = df['Age'].astype(int)
df
```

## Merging / Joining dataframes:

```python
## creation of small dataframe for merging:
data1 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 18, 30, 22, 56],
    "Salary": [100000, 60000, 20000, 45000, 55000],
}
df1 = pd.DataFrame(data1)
df1
```

```python
## creation of small dataframe for merging:
data2 = {
    "Name": ["Alice", "Bob", "Champa", "Dev", "Elen"],
    "Age": [35, 76, 35, 28, 90],
    "Salary": [100500, 60000, 40000, 45000, 30000],
}
df2 = pd.DataFrame(data2)
df2
```

```python
## inner join:
df_inner = pd.merge(df1, df2, on='Name', how='inner')
print("Inner Join:\n")
df_inner
```

```python
## Outer join:
df_outer = pd.merge(df1, df2, on='Name', how='outer')
print("Outer Join:\n")
df_outer
```

```python
## Left join:
df_left = pd.merge(df1, df2, on='Name', how='left')
print("Left Join:\n")
df_left
```

```python
## Right join:
df_right = pd.merge(df1, df2, on='Name', how='right')
print("Right Join:\n")
df_right
```

```python
## Full outer join:
df_full_outer = pd.merge(df1, df2, on='Name', how='outer', indicator=True)
print("Full Outer Join:\n")
df_full_outer
```
